✅ Basic Idea
import numpy as np
🔹 1D Array (Vector)
arr = np.array([1, 2, 3, 4])
print(arr.ndim)

Output:

1

✔ Only one axis → just a list of values

🔹 2D Array (Matrix)
arr = np.array([[1, 2],
                [3, 4]])
print(arr.ndim)

Output:

2

✔ Two axes → rows and columns

🔹 3D Array
arr = np.array([[[1, 2],
                 [3, 4]]])
print(arr.ndim)

Output:

3

✔ Three axes → like stacked matrices

✅ What Does Dimension Actually Mean?
Array Type	Shape Example	ndim
Scalar	5	0
1D Array	(4,)	1
2D Array	(2,2)	2
3D Array	(1,2,2)	3
✅ Difference Between ndim and shape

ndim → tells how many axes

shape → tells size of each axis

Example:

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print(arr.ndim)   # 2
print(arr.shape)  # (2, 3)

✔ 2 dimensions
✔ 2 rows and 3 columns

📌 Difference Between flatten() and ravel() in NumPy
🔹 1️⃣ flatten()
arr.flatten()

Returns a copy

Changes to result do NOT affect original array

Slightly slower (because memory is duplicated)

Example:
import numpy as np

arr = np.array([[1, 2], [3, 4]])
flat = arr.flatten()

flat[0] = 99

print(arr)
print(flat)

Output:

[[1 2]
 [3 4]]

[99  2  3  4]

✔ Original array unchanged.

🔹 2️⃣ ravel()
arr.ravel()

Returns a view (if possible)

Changes to result may affect original array

Faster (no new memory allocation)

Example:
arr = np.array([[1, 2], [3, 4]])
rav = arr.ravel()

rav[0] = 99

print(arr)
print(rav)

Output:

[[99  2]
 [ 3  4]]

[99  2  3  4]

✔ Original array changed.

```python
salaries = np.array([50000, 60000, np.nan, 80000, 120000, np.nan])
```

```python
mean=np.nanmean(salaries)
salaries=np.nan_to_num(salaries,nan=mean)
salaries
```

```python
norm_salaries=(salaries-np.mean(salaries))/np.std(salaries)
```

```python
print(norm_salaries)
```

```python
outliers=salaries[np.abs(norm_salaries)>2]
outliers
```

```python
summary = {
    "mean": np.mean(salaries),
    "median": np.median(salaries),
    "std": np.std(salaries),
    "max": np.max(salaries),
    "min": np.min(salaries)
}
summary
```

```python
data = np.vstack((salaries, norm_salaries))
data
```

```python
cov_matrix = np.cov(data)
cov_matrix
```
